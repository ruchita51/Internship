# 4  Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
req=requests.get("https://meesho.com/sarees/pl/f40kj")
soup=BeautifulSoup(req.text,'html.parser')

product_name=[]
discount=[]
price=[]

for i in soup.find_all(class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    product_name.append(i.text)
    
    
for i in soup.find_all(class_="Card__BaseCard-sc-b3n78k-0 iJCKg NewProductCard__DiscountRow-sc-j0e7tu-15 kUcVjG NewProductCard__DiscountRow-sc-j0e7tu-15 kUcVjG"):
    discount.append(i.text)
    
for i in soup.find_all(class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    price.append(i.text)
    

df=pd.DataFrame({"product_name":product_name, "discount":discount,"price":price})
df
```

# 5  a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
req=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
soup=BeautifulSoup(req.text)
```

```python
rating=[]
points=[]
matches=[]
country=[]
for i in soup.find_all(class_="table-body"):
    country.append(i.text.split()[-5])
    del country[10:30]
for i in soup.find_all(class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
    del rating[10:30]
for i in soup.find_all(class_="table-body"):
    matches.append(i.text.split()[-3])
    del matches[10:30]
for i in soup.find_all(class_="table-body"):
    points.append(i.text.split()[-2])
    del points[10:30]
```

```python
df=pd.DataFrame({"country_name":country,"matches":matches,"ratings":rating,"point":points})
```

```python
df
```

# 5 b) Top 10 ODI Batsmen along with the records of their team and rating

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi'

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
name=[]
for p in soup.find_all(class_="table-body__cell name"):  
    name.append(p.text.split()[0:2])
del name[9:67]
teams=[]
for p in soup.find_all("span",class_="table-body__logo-text"):  
    teams.append(p.text)
del teams[9:67]
posi=[]
for p in soup.find_all("span",class_="rankings-table__pos-number"):  
    posi.append(p.text.split()[0])
del posi[9:67]
posi
rating=[]
for p in soup.find_all("td",class_="table-body__cell u-text-right rating"):  
    rating.append(p.text)
del rating[9:67]
import pandas as pd
df1=pd.DataFrame({"position":posi,"name":name,"country":teams,"rating":rating})
df1
```

```python
name=soup.find(class_="rankings-block__banner--name")
name.text
rating=soup.find("div",class_="rankings-block__banner--rating")
rating.text
team=soup.find("span",class_="rankings-block__pos-number")
team.text.split()[0]
team
nat=[]
for i in soup.find_all(class_="rankings-block__banner--player-info"):
    nat.append(i.text.split()[2])
del nat[1:6]
nat
import pandas as pd
df=pd.DataFrame({"position":team,"name":name,"team":nat,"rating":rating})
df
```

```python
frames = [df, df1]
  
result = pd.concat(frames)
display(result)
```

# 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.bewakoof.com/women-tshirts?ga_q=tshirts'

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

url=[]
name=[]
price=[]

for p in soup.find_all("img",class_="productImgTag"):
    url.append(p["src"])

for p in soup.find_all("h3"):  
    name.append(p.text)
del name[10]

for p in soup.find_all("b"):  
    price.append(p.text)
del price[10:67]
```

```python
import pandas as pd
df=pd.DataFrame({"product_name":name,"price":price,"img_url":url})
df
```
